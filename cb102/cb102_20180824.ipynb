```r
customer_s = scale(customer[,-1])
```

```r
print( mean( customer_s[,2] ) )
print( sd( customer_s[,2] ) )
```

```r
?hclust
```

```r
hc = hclust( dist(customer_s),method = 'complete' )
```

```r
plot(hc,cex=0.8,hang = -0.01)
```

```r
fit = cutree(hc,k=3)
```

```r
fit = cutree(hc,k=4)
```

```r
table(fit)
```

```r
summary(customer[fit == 1,])
```

```r
summary(customer[fit == 2,])
```

```r
hc2 = hclust( dist(customer_s),method = 'single' )
```

```r
plot(hc2,cex=0.8,hang = -0.01)
rect.hclust(hc2,k=4)
```

```r
cutree(hc2,k=3)
```

```r
cutree(hc2,k=4)
```
