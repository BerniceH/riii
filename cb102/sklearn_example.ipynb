```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.data.shape
```

```python
cancer.feature_names
```

```python
cancer.target
```

```python
cancer.target_names
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(cancer.data,cancer.target,test_size=0.2)
```

```python
from sklearn import tree
```

```python
cart_tree  = tree.DecisionTreeClassifier()
```

```python
cart_tree.fit(X_train,Y_train)
```

```python
Y_predict = cart_tree.predict(X_test)
```

```python
Y_predict
```

```python
sum(Y_predict == Y_test) / len(Y_predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_predict,Y_test)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score
```

```python
precision_score(Y_test,Y_predict),recall_score(Y_test,Y_predict),accuracy_score(Y_test,Y_predict)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scoring = {'acc':'accuracy','precision':'precision','recall':'recall'}
```

```python
cross_validate(cart_tree,X_train,Y_train,cv=10,scoring=scoring)
```

```python
from sklearn.metrics import roc_curve,auc
```

```python
fpr,tpr,_ = roc_curve(Y_test,Y_predict)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
```

```python
plt.plot(fpr,tpr,color='darkorange',lw=2,label="ROC curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/YorkLin/lecture/riii/data/customer.csv')
```

```python
df = df.drop('ID',axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import scale
```

```python
df.values
```

```python
type(df.values)
```

```python
customer_s = scale(df.values)
```

```python
from sklearn.cluster import KMeans
```

```python
?KMeans
```

```python
k_model = KMeans(n_clusters=4,n_init=5)
```

```python
cluster = k_model.fit_predict(customer_s)
```

```python
df[cluster == 1]
```

```python
from sklearn.metrics import silhouette_score
```

```python
?silhouette_score
```

```python
sil_ary = list()
for k in range(2,10+1):
    k_model = KMeans(n_clusters=k,n_init=5)
    cluster = k_model.fit_predict(customer_s)
    sil_ary.append({'nk':k,'silhouette':silhouette_score(customer_s,cluster)})
```

```python
%pylab inline
df_ = pd.DataFrame(sil_ary)
df_.plot(x='nk',y='silhouette')
```
