```r
library(ggmap)
```

```r
setwd('~/lecture/riii')
ptt_sample = read.csv('./data/ptt_sample.csv',header=T)
```

```r
head(ptt_sample)
```

```r
# map = get_map(location='Taiwan',zoom=8,language = 'zh-TW')
map = get_map(location='Taiwan',zoom=8,language = 'zh-TW',source='stamen',maptype = 'toner-lite' )
```

```r
g_map = ggmap(map)+geom_count(data=ptt_sample,aes(x=lon,y=lat,color=..n..))+scale_color_gradient(low='yellow',high='red')
```

```r
g_map
```
